###  Insights:
#### 1.	Distribution of dog cuteness
* •	Not all dogs can be the same. Some might be cuter than others or generally preferred by their cuteness level. The levels have also been categorized into four. 
* The data was examined based on the cuteness of the various dogs. This was generated by using the rating column. 
#### 2.	Highly rated dog by name
* •	Some people generally choose or prefer their dogs based on their names making their love for them greater.
* This insight was generated by first grouping the names of the dogs with the ratings. The average was found of the respective dog names in relation to rating. It shows how ratings are considering just dog names.
#### 3.	Highly rated dog by stage
* •	Some people also prefer some dogs based on their stages and rate them accordingly.
* This insight was generated by first grouping the various stages of the dogs with the ratings. The average was found of the respective dog stages in relation to rating. It shows how ratings are considering just dog stages.
